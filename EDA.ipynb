Loading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"D:\python project\ICRISAT-District Level Data (Python).csv")
df = pd.DataFrame(df)
# print(df.shape())
# print(df.head())
df
```

Understanding the Dataset

```python
print("Printing the dataset info \n", df.info(),"\n\n")
print("Null Data: ",df.isnull().sum())
```

Handling the missing data if any is present and Data Cleaning

```python
df.replace(-1,np.nan,inplace = True)
df = df.dropna()
df
```

Pair Plot of the major Crops

```python
cols = [
    "RICE PRODUCTION (1000 tons)",
    "WHEAT PRODUCTION (1000 tons)",
    "MAIZE PRODUCTION (1000 tons)",
    "COTTON PRODUCTION (1000 tons)", 
    "SUGARCANE PRODUCTION (1000 tons)",
]
sample_df = df[cols].dropna().sample(500, random_state=42)
rename_dict = {
    "RICE PRODUCTION (1000 tons)": "Rice Prod",
    "WHEAT PRODUCTION (1000 tons)": "Wheat Prod",
    "MAIZE PRODUCTION (1000 tons)": "Maize Prod",
    "COTTON PRODUCTION (1000 tons)": "Cotton Prod",
    "SUGARCANE PRODUCTION (1000 tons)": "Sugarcane Prod",
}
sample_df = sample_df.rename(columns=rename_dict)
sns.pairplot(sample_df)
plt.suptitle("Pairplot of Crop Production and Yield", y=1.02)
plt.show()
```

```python
df_year_prod = df.groupby("Year")[
    ["RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)",
    "MAIZE PRODUCTION (1000 tons)", "COTTON PRODUCTION (1000 tons)",
    "SUGARCANE PRODUCTION (1000 tons)"]
].sum()
df_heatmap = df_year_prod.T

plt.figure(figsize=(20, 6))
sns.heatmap(df_heatmap, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap of Crop Production by Year")
plt.xlabel("Year")
plt.ylabel("Crop Type")
plt.tight_layout()
plt.show()
```

Rice Production Over years

```python
riceproduction = df.groupby("Year")["RICE YIELD (Kg per ha)"].mean()
plt.figure(figsize=(10,6))
sns.lineplot(x=riceproduction.index, y=riceproduction.values, marker='o')
plt.title("Average Rice Yield Over Years")
plt.xlabel("Year")
plt.ylabel("Yield (Kg per ha)")
plt.grid(True)
plt.show()
```

All Crop production year wise

```python
year = 2015 #You  can give the year and check those year production
subset = df[df["Year"] == year]

prod_cols = [col for col in df.columns if "PRODUCTION" in col and "1000 tons" in col]
total_production = subset[prod_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=total_production.values, y=total_production.index, palette="viridis")
plt.title(f"Crop Production in {year}")
plt.xlabel("Total Production (1000 tons)")
plt.show()
```

Box plot of the major Crops

```python
major_yields = [
    "RICE YIELD (Kg per ha)", 
    "WHEAT YIELD (Kg per ha)",
    "MAIZE YIELD (Kg per ha)",
    "COTTON YIELD (Kg per ha)",
    "SUGARCANE YIELD (Kg per ha)"
]

melted = df.melt(value_vars=major_yields, var_name='Crop', value_name='Yield')

plt.figure(figsize=(12,6))
sns.boxplot(data=melted, x='Crop', y='Yield', palette="Set2", showfliers=False)
plt.title("Yield Distribution of Major Crops")
plt.xticks(rotation=45)
plt.show()
```

Identify High and Low Yeild Regions

```python
state_yield = df.groupby("State Name")[["RICE YIELD (Kg per ha)"]].mean().sort_values("RICE YIELD (Kg per ha)", ascending=False)
top_states = state_yield.head(10)
bottom_states = state_yield.tail(10)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
top_states.plot(kind="barh", ax=ax[0], color="green", legend=False)
bottom_states.plot(kind="barh", ax=ax[1], color="skyblue", legend=False)
ax[0].set_title("Top 10 States by Rice Yield")
ax[1].set_title("Bottom 10 States by Rice Yield")

ax[0].invert_yaxis()
ax[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

Regression relation between Years and total Production

```python
df_year_prod = df.groupby("Year")[
    ["RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)",
    "MAIZE PRODUCTION (1000 tons)", "COTTON PRODUCTION (1000 tons)",
    "SUGARCANE PRODUCTION (1000 tons)"]
].sum()
df_year_prod["Total Production"] = df_year_prod.sum(axis=1)
plt.figure(figsize=(10, 6))
sns.regplot(x=df_year_prod.index, y=df_year_prod["Total Production"])
plt.title("Year vs Total Crop Production")
plt.xlabel("Year")
plt.ylabel("Total Production (1000 tons)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Scatter plot between Rice area and it's Production

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, 
    x="RICE AREA (1000 ha)", 
    y="RICE PRODUCTION (1000 tons)", 
    hue="Year", palette="coolwarm", alpha=0.6
)
plt.title("Rice Area vs Production")
plt.xlabel("Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Major crop production in the country

```python
total_crop_share = {
    "Rice": df["RICE PRODUCTION (1000 tons)"].sum(),
    "Wheat": df["WHEAT PRODUCTION (1000 tons)"].sum(),
    "Maize": df["MAIZE PRODUCTION (1000 tons)"].sum(),
    "Cotton": df["COTTON PRODUCTION (1000 tons)"].sum(),
    "Sugarcane": df["SUGARCANE PRODUCTION (1000 tons)"].sum()
}

plt.figure(figsize=(8, 8))
plt.pie(total_crop_share.values(), labels=total_crop_share.keys(), autopct="%1.1f%%", startangle=140)
plt.title("Crop-wise Production Share in India")
plt.axis("equal")
plt.show()
```
